# How to use UNESCAP Carbon Pricing Instrument (CPI) model

## Starting 

The tool is started by this [link](https://mybinder.org/v2/gh/IbHansen/UNESCAP-Asia/HEAD?urlpath=tree/Asia/Carbon%20tax%20experiments%20singel%20country.ipynb)



The link will open a page in your browser which looks like this: 

It can take some time for the tool to apear. First a virtual machine is created, then the model has to be loaded. 

## Input 

Then a country has to be selected from the list of countries. 

When a country has been selected, Three input tabs will be displayed. They can be expanded in order to input different policy options. 



## Simulating

Now the model can be simulated by pressing this button: 



Before simulating a name can be assigned instead of the suggested by filling out this field: 


## Inspect output 

When the model is solved the output can be inspected. Different indicators can be selected from this dropdown box: 



The format can be adjusted from this control panel: 



For some indicators it can be more relevant to look at the difference to "Business as usual" either in absolute terms or in percent of "Business as usual"

Also for some indicators it can be helpful to toggle Legends off. 

## Restart a country
To restart a country, just select another country and select the same country again.
